```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
T = 1.0            # total time
N = 1000           # number of steps
dt = T / N         # time step
t = np.linspace(0, T, N+1)  # time grid
```

```python
# Simulate increments
dW = np.random.normal(0, np.sqrt(dt), size=N)  # N increments ~ N(0, dt)
W = np.zeros(N+1)                              # initialize Brownian path
W[1:] = np.cumsum(dW)                          # cumulative sum to get W(t)
```

```python
# Plot the path
plt.figure(figsize=(10, 4))
plt.plot(t, W, label="Wiener Process Path")
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.title("Simulated Path of Standard Brownian Motion")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Output the full path
print("Time points:", t)
print("Wiener process path:", W)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
T = 1.0        # Total time
N = 1000       # Number of time steps
dt = T / N     # Time step
t = np.linspace(0, T, N+1)  # Time grid
```

```python
S0 = 100       # Initial value of the process
alpha = 0.1    # Drift coefficient
sigma = 0.2    # Volatility coefficient
```

```python
# Simulate 5 paths
paths = np.zeros((5, N+1))

for i in range(5):
    dW = np.random.normal(0, np.sqrt(dt), size=N)
    W = np.cumsum(dW)
    W = np.insert(W, 0, 0)  # Insert W(0) = 0
    S = S0 * np.exp(sigma * W + (alpha - 0.5 * sigma**2) * t)
    paths[i] = S
```

```python
# Plot the paths
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.plot(t, paths[i], label=f'Path {i+1}')
plt.xlabel('Time t')
plt.ylabel('S(t)')
plt.title('5 Simulated Paths of Geometric Brownian Motion')
plt.grid(True)
plt.legend()
plt.show()
```
